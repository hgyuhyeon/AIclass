```python
# 최종 Master, Miss, Mr, Mrs, Other 로 변경 예정 : Master 어린 아이에게 붙이는 이니셜
before = ['Capt', 'Col', 'Countess', 'Don', 'Dona',
           'Dr', 'Jonkheer', 'Lady', 'Major', 'Mlle',
           'Mme', 'Ms', 'Rev', 'Sir']
after = ['Mr', 'Mr', 'Mrs', 'Mr', 'Miss',
         'Other', 'Mr', 'Mrs', 'Mr', 'Miss',
         'Mrs', 'Miss', 'Other', 'Mr']
df['initial'].replace(before, after, inplace=True)
```

```python
# X
#        [선실등급, 성별, 나이, 형제자매_수, 부모자식_수, 운임]
x_cols = [ 'pclass', 'sex', 'age', 'sibsp', 'parch', 'fare' ]
X = df[ x_cols ].values   # numpy array 로 리턴
X
```

```python
# y
y = df['survived'].values
y
```

```python
##### 데이터 전처리 완료!!!
```

```python
# ML 코딩 5단계
# 1. 훈련 데이터와 테스트 데이터로 분리 : (X_train, X_test), (y_train, y_test)
# 2. 머신러닝 알고리즘 선택 : 예) DecisionTreeClassifier
# 3. 훈련 : model.fit(X_train, y_train)
# 4. 점수 : model.score(X_test, y_test)
# 5. 새 데이터로 예측 : 예) 디카프리오 정보로 생존 예측
```

```python
# 1. 훈련 데이터와 테스트 데이터로 분리 : (X_train, X_test), (y_train, y_test)
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42) 
# 테스트 데이터 10%, random_state: 재현성 보장. 같은 결과를 내줌
```

```python
# 2. 머신러닝 알고리즘 선택 : 예) DecesionTreeClassifier
from sklearn.tree import DecisionTreeClassifier #Classifier: 분류, Regressor: 회귀
tree_clf = DecisionTreeClassifier(max_depth=5) # 생성자를 통해 객체를 만든다.
#max_depth = 하이퍼파라미터, 이 값을 설정하면서 오버피팅을 조절함
```

```python
# 3. 훈련 : model.fit(X_train, y_train)
tree_clf.fit(X_train, y_train) # 훈련할 X와 정답 y를 가지고 학습
```

```python
# 4. 점수 : model.score(X_test, y_test)
tree_clf.score(X_train, y_train) # 훈련 점수
```

```python
tree_clf.score(X_test, y_test) # 테스트 점수
```

```python
# 5. 새 데이터로 예측 : 예) 디카프리오 정보로 생존 예측
```

```python
#        [선실등급: 3등실, 성별: 남자, 나이; 19, 형제자매_수: 0, 부모자식_수: 0, 운임: 10$]
#        [선실등급: 1등실, 성별: 여자, 나이; 17, 형제자매_수: 1, 부모자식_수: 2, 운임: 100$]
testset = [[3, 0, 19, 0, 0, 10 ], [1, 1, 17, 1, 2, 100]] # 1개의 테스트를 하더라도 꼭 2차원 데이터 생성

is_survived = tree_clf.predict(testset)
print(is_survived) # [0] => 사망
result = tree_clf.predict_proba(testset) # %로 표기됨
print(f"디카프리오 생존 확률: {result[0][1] * 100:.2f}%, 사망 확률: {result[0][0] * 100:.2f}%") # 사망확률 85%, 생존확률 14%
print(f"윈슬릿 생존 확률: {result[1][1] * 100:.2f}%, 사망 확률: {result[1][0] * 100:.2f}%") # 사망확률 85%, 생존확률 14%
```
